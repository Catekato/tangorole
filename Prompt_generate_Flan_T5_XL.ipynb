### Load the model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xl")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xl").to("cuda")
```

```python
import json
```

```python
file1_path = '/content/all_test_original.json'
with open(file1_path, 'r', encoding='utf-8') as f1:
    data1 = json.load(f1)

output_path = '/content/empty.json'
captions = []
texts = []
for i in range(len(data1)):
        captions.append(data1[i]['captions'])
```

```python
def generate(input_text):
  input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")
  output = model.generate(input_ids, max_length=100)
  return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
for caption in captions:
  input_text = f"""You’re a text-to-audio generation model user, give a more global description of the sound in the {caption}"""
  text= generate(input_text)
  texts.append(text)
```

```python
output_path = '/content/empty.json'
with open(output_path, 'w', encoding='utf-8') as output_file:
  json.dump(texts, output_file, ensure_ascii=False, indent=4)
```
